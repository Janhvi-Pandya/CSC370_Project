```python
import mysql.connector
from mysql.connector import errorcode

def get_db_connection(user="root", password="root"):
    try:
        conn = mysql.connector.connect(
            host="localhost",
            user=user,
            password=password,
            database="educonnect_demo"
        )
        return conn
    except mysql.connector.Error as err:
        if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
            print("Something is wrong with your user name or password")
        elif err.errno == errorcode.ER_BAD_DB_ERROR:
            print("Database does not exist")
        else:
            print(err)
        return None

def execute_query(query, params=(), fetch=False, user="root", password="root"):
    conn = get_db_connection(user=user, password=password)
    if conn:
        try:
            cursor = conn.cursor()
            print(f"Executing SQL: {query}")
            cursor.execute(query, params)
            if fetch:
                results = cursor.fetchall()
            else:
                results = None
            conn.commit()
            cursor.close()
            conn.close()
            return results
        except mysql.connector.Error as err:
            print(err)
            return None
    else:
        print("Failed to establish a database connection.")
        return None


def grouping_and_aggregation():
    query = """
    SELECT UserID, COUNT(ProjectID) AS ProjectCount
    FROM Projects
    GROUP BY UserID;
    """
    results = execute_query(query, fetch=True)
    if results:
        for row in results:
            print(row)

def sql_dml_and_fks():
    query = """
    INSERT INTO Users (Name, Email, Password, Bio, ProfilePic)
    VALUES ('John Doe', 'john.doe.unique2@example.com', 'securepassword', 'Tech enthusiast', 'john.jpg');
    """
    execute_query(query)
    print("User John Doe inserted successfully.")

def sub_queries():
    query = """
    SELECT Name, Email
    FROM Users
    WHERE UserID IN (
        SELECT UserID
        FROM Projects
        GROUP BY UserID
        HAVING COUNT(ProjectID) > 1
    );
    """
    results = execute_query(query, fetch=True)
    if results:
        for row in results:
            print(row)

def using_indexes():
    query = "CREATE INDEX idx_users_email ON Users (Email);"
    execute_query(query)
    print("Index idx_users_email created successfully.")

def basic_security():
    query = """
    SELECT UserID, Name
    FROM Users
    WHERE Email = 'alice@example.com' AND Password = 'password123';
    """
    results = execute_query(query, fetch=True)
    if results:
        for row in results:
            print(row)

def managing_access():
    pass;



def understanding_views():
    execute_query("DROP VIEW IF EXISTS UserProjectCounts;")
    query = """
    CREATE VIEW UserProjectCounts AS
    SELECT Users.UserID, Users.Name, COUNT(Projects.ProjectID) AS ProjectCount
    FROM Users
    LEFT JOIN Projects ON Users.UserID = Projects.UserID
    GROUP BY Users.UserID, Users.Name
    """
    execute_query(query)
    print("View UserProjectCounts created successfully.")
    query = "SELECT * FROM UserProjectCounts WHERE ProjectCount > 1;"
    results = execute_query(query, fetch=True)
    if results:
        for row in results:
            print(row)

def acid_transactions():
    conn = get_db_connection()
    if conn:
        cursor = conn.cursor()
        try:
            cursor.execute("START TRANSACTION;")
            cursor.execute("""
            INSERT INTO Users (Name, Email, Password, Bio, ProfilePic)
            VALUES ('Jane Doe', 'jane.doe.unique2@example.com', 'anotherpassword', 'Bio for Jane Doe', 'jane.jpg');
            """)
            cursor.execute("""
            INSERT INTO Projects (UserID, Title, Description, Category, Tags)
            VALUES ((SELECT UserID FROM Users WHERE Email = 'jane.doe.unique2@example.com'), 'Janes Project', 'Description of Janes Project', 'Science', 'Research, Innovation');
            """)
            conn.commit()
            print("Transaction committed successfully.")
        except Exception as e:
            conn.rollback()
            print("Transaction rolled back due to error:", e)
        finally:
            cursor.close()
            conn.close()
    else:
        print("Failed to establish a database connection.")

if __name__ == "__main__":
    print("Grouping & Aggregation")
    grouping_and_aggregation()
    print("\nSQL DML & FKs")
    sql_dml_and_fks()
    print("\nSub-queries")
    sub_queries()
    print("\nUsing Indexes")
    using_indexes()
    print("\nBasic Security")
    basic_security()
    print("\nManaging Access")
    managing_access()
    print("\nUnderstanding Views")
    understanding_views()
    print("\nACID Transactions")
    acid_transactions()
```

```python
import mysql.connector
from mysql.connector import errorcode

def get_db_connection(user="root", password="root"):
    try:
        conn = mysql.connector.connect(
            host="localhost",
            user=user,
            password=password,
            database="educonnect_demo"
        )
        return conn
    except mysql.connector.Error as err:
        if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
            print("Something is wrong with your user name or password")
        elif err.errno == errorcode.ER_BAD_DB_ERROR:
            print("Database does not exist")
        else:
            print(err)
        return None

def execute_query(query, params=(), fetch=False, user="root", password="root"):
    conn = get_db_connection(user=user, password=password)
    if conn:
        try:
            cursor = conn.cursor()
            cursor.execute(query, params)
            if fetch:
                results = cursor.fetchall()
            else:
                results = None
            conn.commit()
            cursor.close()
            conn.close()
            return results
        except mysql.connector.Error as err:
            print(err)
            return None
    else:
        print("Failed to establish a database connection.")
        return None

def show_privileges(user):
    query = f"SHOW GRANTS FOR '{user}'@'localhost';"
    results = execute_query(query, fetch=True)
    if results:
        for row in results:
            print(row)

def display_users():
    query = "SELECT User, Host FROM mysql.user;"
    results = execute_query(query, fetch=True)
    if results:
        for row in results:
            print(row)

def display_custom_users():
    query = "SELECT UserID, Name, Email FROM Users;"
    results = execute_query(query, fetch=True)
    if results:
        for row in results:
            print(row)

def grant_privileges():
    queries = [
        "CREATE ROLE IF NOT EXISTS 'read_only';",
        "CREATE ROLE IF NOT EXISTS 'read_write';",
        "GRANT SELECT ON educonnect_demo.* TO 'read_only';",
        "GRANT SELECT, INSERT, UPDATE, DELETE ON educonnect_demo.* TO 'read_write';",
        "CREATE USER IF NOT EXISTS 'readonly_user'@'localhost' IDENTIFIED BY 'readonly_password';",
        "CREATE USER IF NOT EXISTS 'readwrite_user'@'localhost' IDENTIFIED BY 'readwrite_password';",
        "GRANT 'read_only' TO 'readonly_user'@'localhost';",
        "GRANT 'read_write' TO 'readwrite_user'@'localhost';",
    ]
    for query in queries:
        execute_query(query)
    print("Roles created and privileges granted successfully.")

def revoke_privileges():
    queries = [
        "REVOKE 'read_only' FROM 'readonly_user'@'localhost';",
        "REVOKE 'read_write' FROM 'readwrite_user'@'localhost';",
        "DROP USER IF EXISTS 'readonly_user'@'localhost';",
        "DROP USER IF EXISTS 'readwrite_user'@'localhost';",
    ]
    for query in queries:
        execute_query(query)
    print("Privileges revoked and users dropped successfully.")

def create_views():
    queries = [
        "DROP VIEW IF EXISTS ReadOnlyProjects;",
        "DROP VIEW IF EXISTS ReadWriteProjects;",
        "CREATE VIEW ReadOnlyProjects AS SELECT ProjectID, UserID, Title, Category, Tags, DateCreated, LastUpdated FROM Projects;",
        "CREATE VIEW ReadWriteProjects AS SELECT * FROM Projects;",
    ]
    for query in queries:
        execute_query(query)
    print("Views created successfully.")

def query_views():
    query = "SELECT * FROM ReadOnlyProjects;"
    results = execute_query(query, fetch=True)
    if results:
        for row in results:
            print(row)

def insert_into_view():
    query = """
    INSERT INTO ReadWriteProjects (UserID, Title, Description, Category, Tags)
    VALUES (21, 'New Project', 'This is a description of the new project.', 'Software Development', 'New, Project');
    """
    execute_query(query)
    print("Data inserted into view successfully.")

def delete_from_view():
    # First delete related comments to avoid foreign key constraint issues
    execute_query("DELETE FROM Comments WHERE ProjectID = 1;")
    query = """
    DELETE FROM ReadWriteProjects
    WHERE ProjectID = 1;
    """
    execute_query(query)
    print("Data deleted from view successfully.")

if __name__ == "__main__":
    print("Initial State: Showing Privileges for root")
    show_privileges('root')

    print("\nInitial State: Displaying Users")
    display_users()
    display_custom_users()

    print("\nCreating Roles and Granting Privileges")
    grant_privileges()

    print("\nAfter Granting Privileges: Showing Privileges for root")
    show_privileges('root')

    print("\nAfter Granting Privileges: Displaying Users")
    display_users()
    display_custom_users()

    print("\nRevoking Privileges")
    revoke_privileges()

    print("\nAfter Revoking Privileges: Showing Privileges for root")
    show_privileges('root')

    print("\nAfter Revoking Privileges: Displaying Users")
    display_users()
    display_custom_users()

    print("\nCreating Views")
    create_views()

    print("\nInitial State: Querying Views")
    query_views()

    print("\nInserting into View")
    insert_into_view()

    print("\nAfter Inserting into View: Querying Views")
    query_views()

    print("\nDeleting from View")
    delete_from_view()

    print("\nAfter Deleting from View: Querying Views")
    query_views()
```
